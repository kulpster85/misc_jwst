# Querying JWST visit status

The JWST Program Information page at https://www.stsci.edu/jwst/science-execution/program-information.html allows querying the status of an observing program. The output is a set of tables that display the status of each visit in that program. 

This notebook demonstrates the use of some functions to retrieve the outputs of that form within Python code. 


The primary interface is the function `query_program_status_form`. This takes a PID as input, and returns
a pandas dataframe table of the outputs. 

```python
vs.query_program_status_form(1138)
```

```python
vs.query_program_status_form(1411)
```
